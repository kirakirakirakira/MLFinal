```python
import matplotlib.pyplot as plt
import numpy as np

# 算法名称
algorithms = ['lstm', 'transformer', 'CSTF']
x = np.arange(len(algorithms))  # x轴刻度位置

# 每组的误差数据（示例数值，请根据你的实际数据替换）
mse_96 = [5.798, 6.603, 8.458]
mae_96 = [0.375, 0.258, 0.522]
mse_240 = [1.869, 4.362, 11.919]
mae_240 = [0.116, 0.287, 0.494]

# 柱状图宽度
width = 0.18

# 绘制柱状图
plt.figure(figsize=(8, 6))
plt.bar(x - 1.5*width, mse_96, width, label='MSE(Output=90)', color='blue')
plt.bar(x - 0.5*width, mae_96, width, label='MAE(Output=90)', color='red')
plt.bar(x + 0.5*width, mse_240, width, label='MSE(Output=365)', color='khaki')
plt.bar(x + 1.5*width, mae_240, width, label='MAE(Output=365)', color='green')

# 图形美化
plt.xticks(x, algorithms)
plt.ylabel('Error')
plt.xlabel('Algorithms')
plt.title('Error Comparison')
plt.legend()
plt.grid(axis='y', linestyle='--', alpha=0.5)
plt.tight_layout()

# 显示图像
plt.show()
```

```python
import json
import matplotlib.pyplot as plt

# 读取 JSON 文件并返回列表
def load_json(file_path):
    with open(file_path, 'r') as f:
        return json.load(f)

# 加载数据
y_lstm = load_json('lstm-90.json')
y_transformer = load_json('tf-90.json')
y_sktnet = load_json('new-90.json')
y_true = load_json('true_val.json')

# 取四个列表的最小长度
min_len = min(len(y_lstm), len(y_transformer), len(y_sktnet), len(y_true))

# 截断所有列表为最短长度
y_lstm = y_lstm[:min_len]
y_transformer = y_transformer[:min_len]
y_sktnet = y_sktnet[:min_len]
y_true = y_true[:min_len]

# 绘制对比曲线
plt.figure(figsize=(10, 6))
plt.plot(y_true, label='Ground Truth', color='red')
plt.plot(y_lstm, label='LSTM', color='blue')
plt.plot(y_transformer, label='Transformer',color='green')
plt.plot(y_sktnet, label='CSTF', color='yellow')

# 图表美化
plt.xlabel('Time Step')
plt.ylabel('Value')
plt.title('Prediction Comparison with Ground Truth(Output=90)')
plt.legend()
plt.grid(True)
plt.tight_layout()

# 显示图像
plt.show()
```
